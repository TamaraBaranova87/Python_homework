```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('horse_data.csv', names=['surgery?','age','hospital Number','rectal temperature','pulse','respiratory rate',
                                          'temperature of extremities','peripheral pulse','mucous membranes',
                                          'capillary refill time','pain','peristalsis','abdominal distension',
                                          'nasogastric tube','nasogastric reflux','nasogastric reflux PH',
                                          'rectal examination - feces','abdomen','packed cell volume','total protein',
                                          'abdominocentesis appearance','abdomcentesis total protein','outcome',
                                         'surgical lesion?','type of lesion1','type of lesion2','type of lesion3','cp_data'])
df = df.replace('?', np.nan)
df = df.replace('', np.nan)
df['surgery?'] = df['surgery?'].replace(['1','2'],['Yes','No'])
df['temperature of extremities'] = df['temperature of extremities'].replace(['1','2','3','4'],['Normal','Warm','Cool','Cold'])
df['peripheral pulse'] = df['peripheral pulse'].replace(['1','2','3','4'],['normal','increased','reduced','absent'])
df['mucous membranes'] = df['mucous membranes'].replace(['1','2','3','4','5','6'],['normal pink','bright pink',
                                                                                   'pale pink','pale cyanotic',
                                                                                   'bright red / injected','dark cyanotic'])
df['capillary refill time'] = df['capillary refill time'].replace(['1','2','3'],['< 3 seconds','>= 3 seconds','>= 3 seconds'])
df['pain'] = df['pain'].replace(['1','2','3','4','5'],['alert, no pain','depressed',
                                                                                   'intermittent mild pain',
                                                           'intermittent severe pain','continuous severe pain'])
df['peristalsis'] = df['peristalsis'].replace(['1','2','3','4'],['hypermotile','normal','hypomotile','absent'])
df['abdominal distension'] = df['abdominal distension'].replace(['1','2','3','4'],['none','slight','moderate','severe'])
df['nasogastric tube'] = df['nasogastric tube'].replace(['1','2','3'],['none','slight','significant'])
df['nasogastric reflux'] = df['nasogastric reflux'].replace(['1','2','3'],['none','> 1 liter','< 1 liter'])
df['abdominocentesis appearance'] = df['abdominocentesis appearance'].replace(['1','2','3'],['clear','cloudy','serosanguinous'])
df['rectal examination - feces'] = df['rectal examination - feces'].replace(['1','2','3','4'],['normal','increased','decreased',
                                                                                               'absent'])
df['abdomen'] = df['abdomen'].replace(['1','2','3','4','5'],['normal','other','firm feces in the large intestine',
                                                             'distended small intestine',
                                                                                   'distended large intestine'])
df['outcome'] = df['outcome'].replace(['1','2','3'],['lived','died','was euthanized'])
df['surgical lesion?'] = df['surgical lesion?'].replace([1,2],['Yes','No'])
df['cp_data'] = df['cp_data'].replace([1,2],['Yes','No'])
df['age'] = df['age'].replace([1,9],['Adult horse','Young (< 6 months)'])
df.head(30)
```

```python
df['rectal temperature'] = pd.to_numeric(df['rectal temperature'], errors='coerce')
df['pulse'] = pd.to_numeric(df['pulse'], errors='coerce')
df['respiratory rate'] = pd.to_numeric(df['respiratory rate'], errors='coerce')
df['packed cell volume'] = pd.to_numeric(df['packed cell volume'], errors='coerce')
df['total protein'] = pd.to_numeric(df['total protein'], errors='coerce')
df['nasogastric reflux PH'] = pd.to_numeric(df['nasogastric reflux PH'], errors='coerce')
df['abdomcentesis total protein'] = pd.to_numeric(df['abdomcentesis total protein'], errors='coerce')
df['hospital Number'] = df['hospital Number'].astype('category')
df['surgery?'] = df['surgery?'].astype('category')
df['age'] = df['age'].astype('category')
df[['temperature of extremities','peripheral pulse','mucous membranes','capillary refill time','pain','peristalsis',
    'abdominal distension','nasogastric tube','nasogastric reflux','abdominocentesis appearance','rectal examination - feces',
    'abdomen','outcome','surgical lesion?','type of lesion1','type of lesion2','type of lesion3','cp_data']] = df[[
    'temperature of extremities','peripheral pulse','mucous membranes',
                                                             'capillary refill time','pain','peristalsis',
                                                             'abdominal distension','nasogastric tube','nasogastric reflux',
                                                             'abdominocentesis appearance','rectal examination - feces',
                                                             'abdomen','outcome','surgical lesion?','type of lesion1',
    'type of lesion2','type of lesion3',
                                                             'cp_data']].astype('category')
```

В основном болеют взрослые лошади. У них наблюдается слабая перистальтика, вздутый живот, очень холодные конечности, периодические слабые боли,отсутствует стул. Лечатся колики путем хирургического вмешательства с осложнениями после него. 60% лошадей выживают.

```python
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['pulse'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers.info()
```

```python
frame = pd.concat([df, remove_outliers]).drop_duplicates(keep=False)

frame[frame['pulse'].notnull()].transpose()
```

```python
#Если были хирургические осложнения, то была операция
counts_group = {}
val = []
for i in df['surgical lesion?']:
    if i not in val:
        val.append(i)
for j in val: 
    counts_ = {}
    for i in df['surgery?'][df['surgical lesion?'] == j]:
            
        if i not in counts_ :
            counts_[i] = 1
        else:
            counts_[i] += 1
    counts_group[j] = counts_
print(counts_group)

for k,v in counts_group.items():
    max_ = 0
    mode_ = None
    for i,j in v.items():
        if max_ < j and str(i) != ('nan'):
            max_ = j
            mode_ = i
    counts_group[k] = mode_
print(counts_group)

mask = df['surgery?'].astype('str').isin(['NaN','nan'])
df.loc[mask, 'surgery?'] = df.loc[mask, 'surgical lesion?'].map(counts_group)  
        
        
df['surgery?'].value_counts()
```
