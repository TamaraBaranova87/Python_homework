### Задание 1

```python
data = pd.read_csv('visit_log.csv', sep=';')
data.head()
```

```python
list1 = ['yandex','google']
list2 = ['paid','email']
for i in list1:
    data.loc[data.traffic_source.str.contains(i), 'source_type'] = 'organic'
for i in list2:
    data.loc[(data.traffic_source.str.contains(i)) & (data.region == 'Russia'), 'source_type'] = 'ad'
    data.loc[(data.traffic_source.str.contains(i)) & (data.region != 'Russia'), 'source_type'] = 'other'
data.loc[pd.isnull(data.source_type), 'source_type'] = data['traffic_source']
data.head(10)
```

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
data.head()
```

```python
from pymystem3 import Mystem
```

```python
m = Mystem()
def lem (row):
    lemmas = m.lemmatize(row['keyword'])
    return ' '.join(lemmas)
```

```python
data['lemmas'] = data.apply(lem, axis = 1)
```

### Задание 4

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
joined = client_base.merge(rzd, on='client_id', how='left')
joined = joined.merge(auto, on='client_id', how='left')
joined = joined.merge(air, on='client_id', how='left')
joined.head()
```

```python
import numpy as np
pivot = pd.pivot_table(joined, index =  'client_id', fill_value = 0, margins = True)
pivot.head()

```
