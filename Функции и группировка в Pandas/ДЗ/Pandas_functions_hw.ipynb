### Задание 1

```python
rating = pd.read_csv('ratings.csv')
rating.head()
```

```python
def class_(row):
    if row['rating'] <= 2:
        res = 'Низкий рейтинг'
    elif row['rating'] <= 4:
        res = 'Средний рейтинг'
    else:
        res = 'Высокий рейтинг'
    return res
```

```python
rating['class'] = rating.apply(class_, axis=1)
```

```python
rating['class'].value_counts()
```

### Задание 2

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
def region (row):
    geo_data = {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток'
        : ['владивосток', 'сахалин', 'хабаровск']}
    d = []
    for i in geo_data.values():
        for j in i:
            d.append(j)
    res = ''
    for k,i in enumerate(d):
        if len(i) <= 4:
            if ((' ' + str(i)[:(len(i)-1)]) in str(row['keyword']).lower()) or (str(i)[:(len(i)-1)] in str(row['keyword']).lower()[:(len(i)-1)]):
                for key,value in geo_data.items():
                    if i in value:
                        res = key
                        return res
            elif k < (len(d) - 1) and res == '':
                continue
            else:
                res = 'undefined'
        else:
            if ((' ' + str(i)[:(len(i)-2)]) in str(row['keyword']).lower()) or (str(i)[:(len(i)-2)] in str(row['keyword']).lower()[:(len(i)-2)]):
                for key,value in geo_data.items():
                    if i in value:
                        res = key
                        return res
            elif k < (len(d) - 1) and res == '':
                continue
            else:
                res = 'undefined'
    return res
```

```python
keywords['region'] = keywords.apply(region, axis=1)
```

```python
keywords.head()
```

```python
keywords['region'].value_counts()
```

```python
keywords[(keywords['region'] == 'Центр')].head(30)                                            
```

### Задание 3

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
years = []
i = 1950
while i < 2011:
    years.append(i)
    i += 1
print(years)
```

```python
def production_year (row):
    res = 0
    for k,i in enumerate(years):
        if str(i) in row['title']:
            res = i
            return res
        elif k < (len(years) - 1) and res == 0:
            continue
        else:
            res = 1900
    return res
```

```python
movies['year'] = movies.apply(production_year, axis=1)
```

```python
movies.head()
```

```python
movies['year'].value_counts()
```

```python
movies[(movies['year'] == 1900)].head(30)     
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
results = joined.groupby('year').agg({'rating': 'mean'})
results.sort_values('rating', ascending=False).head()
```
