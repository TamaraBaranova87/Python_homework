### Задание 1

```python
import requests
```

```python
import pandas as pd
```

```python
res = requests.get('https://habr.com/ru/all/')
```

```python
KEYWORDS = ['python', 'парсинг']
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
posts = soup.find_all('article', class_='post')
posts
```

```python
date_ = []
ref = []
title = []
for post in posts:
    post_id = post.parent.attrs.get('id')
    if not post_id:
        continue
    post_id = int(post_id.split('_')[-1])
    hubs = post.find_all('a', class_='hub-link')
    for hub in hubs:
            hub_lower = hub.text.lower()
            if any([desired in hub_lower for desired in KEYWORDS]):
                title_element = post.find('a', class_='post__title_link')
                date_element =  post.find('span', class_='post__time')
                date_.append(date_element.text)
                title.append(title_element.text)
                ref.append(title_element.attrs.get('href'))
                    
                break
inf = {'Наименование статьи':title,'Ссылка':ref}
pd.DataFrame(data = inf, index = date_) 
```

```python

date_ = []
text_ =[]
ref = []
title = []
for post in posts:
    post_id = post.parent.attrs.get('id')
    if not post_id:
        continue
    post_id = int(post_id.split('_')[-1])
    hubs = post.find_all('a', class_='hub-link')
    
    for hub in hubs:
            hub_lower = hub.text.lower()
            texts = post.find('div', class_='post__text').text.lower()
           
            if any([(desired in hub_lower) or (desired in texts) for desired in KEYWORDS]):
                    title_element = post.find('a', class_='post__title_link')
                    date_element = post.find('span', class_='post__time')
                    text_element = post.find('div', class_='post__text')
                   
                    date_.append(date_element.text)
                    title.append(title_element.text)
                    ref.append(title_element.attrs.get('href'))
                    text_.append(text_element.text)
                    break
inf = {'Наименование статьи':title,'Ссылка':ref,'Текст':text_}
pd.DataFrame(data = inf, index = date_)                  
             
```

```python
res = requests.post(url, json={"emailAddresses": emails}, headers={"Vaar-Version": "0", "Vaar-Header-App-Product":  "hackcheck-web-avast"})
```

```python
res.json()
```

```python
emails_dic = {}
emails__ = []
for email,breaches in res.json()['summary'].items():
    emails_dic[email] = breaches['breaches']
    
breaches_site = []
breaches_date = []
breaches_descr = []
for email,breaches in emails_dic.items():
    for i in breaches:
        for key, value in res.json()['breaches'].items():
            
            if value['breachId'] == i:
                breaches_site.append(value['site'])
                breaches_date.append(value['publishDate'])
                breaches_descr.append(value['description'])
                emails__.append(email)
breaches_dic = {'email': emails__,'publishDate':breaches_date,'site':breaches_site,'description':breaches_descr}
pd.DataFrame(data = breaches_dic)       
```
