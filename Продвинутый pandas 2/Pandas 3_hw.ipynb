```python
dt = pd.read_csv('ml-latest-small/ratings.csv')
dt.head()
```

```python
dt_grouped = dt.groupby('userId').agg(['min','max','count'])['timestamp'].sort_values(by='count', ascending=True)
dt_grouped['diff'] = dt_grouped['max'] - dt_grouped['min']
dt_group = dt_grouped.query('count > 100') 
dt_group.head()
```
