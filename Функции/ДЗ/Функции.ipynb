```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
while True:
    user_input = input('\nВведите команду\n')
    if user_input == 'p':
        owner()
    elif user_input == 's':
        shelf()
    elif user_input == 'l':
        inf()
    elif user_input == 'ads':
        add_shelf()
    elif user_input == 'ds':
        del_shelf()
    elif user_input == 'ad':
        add_doc()
    elif user_input == 'd':
        del_doc()
    elif user_input == 'm':
        mov_doc()
    elif user_input == 'q':
        break
```

```python
def owner():
    """
    Пользователь по команде может узнать владельца документа по его номеру
    """
    x = input('\nВведите номер документа:\n')
    y = 0
    for document in documents:
        if document['number'] == x:
            y = document['name']
            print('\nВладелец документа: ' + y)
            break
    if y == 0:
        print('\nДокумент не найден в базе')
    return  
```

```python
def shelf():
    """
    Пользователь по команде может по номеру документа узнать, на какой полке он хранится
    """
    x = input('\nВведите номер документа:\n')
    y = 0
    for key,value in directories.items():
        if x in value:
            y = key
            print('\nДокумент хранится на полке: ' + y)
            break
    if y == 0:
        print('\nДокумент не найден в базе')
    return
```

```python
def inf():
    """
    Пользователь по команде может увидеть полную информацию по всем документам
    """
    for document in documents:
        for key,value in directories.items():
            if document['number'] in value:
                print('\n№: ' + document['number'] + ', тип: ' + document['type'] + 
                         ', владелец: ' + document['name'] + ', полка хранения: ' + key)
    return
```

```python
def add_shelf():
    """
    Пользователь по команде может добавить новую полку
    """
    x = input('\nВведите номер полки:\n')
    if x in list(directories.keys()):
        print('Такая полка уже существует. Текущий перечень полок: ' + ', '.join(list(directories.keys())))
    else:            
        directories[x]=[]
        print('Полка добавлена. Текущий перечень полок: ' + ', '.join(list(directories.keys())))
    return   
```

```python
def del_shelf():
    """
    Пользователь по команде может удалить существующую полку из данных (только если она пустая)
    """
    x = input('\nВведите номер полки:\n')
    if x in list(directories.keys()):
        if directories[x] == []:
            del directories[x]
            print('Полка удалена. Текущий перечень полок: ' + ', '.join(list(directories.keys())))
        else:
            print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ' 
                  + ', '.join(list(directories.keys())))
    else:            
        print('Такой полки не существует. Текущий перечень полок: ' + ', '.join(list(directories.keys())))
    return
```

```python
def add_doc():
    """
    Пользователь по команде может добавить новый документ в данные
    """
    x = input('\nВведите номер документа:\n')
    y = input('\nВведите тип документа:\n')
    z = input('\nВведите владельца документа:\n')
    k = input('\nВведите полку для хранения:\n')
    if k in list(directories.keys()):
        document = {}
        documents.append(document)
        document['type'] = y
        document['number'] = x
        document['name'] = z
        directories[k].append(x)
        print('\nДокумент добавлен. Текущий список документов:')
    else:
        print('\nТакой полки не существует. Добавьте полку командой ads.')
        print('Текущий список документов:')
    inf()
    return
```

```python
def del_doc():
    """
    Пользователь по команде может удалить документ из данных
    """
    x = input('\nВведите номер документа:\n')
    y = 0
    for document in documents:
        if document['number'] == x:
            y = document['number']
            documents.remove(document)
            for k in directories.values():
                if x in k:
                    k.remove(x)
            print('\nДокумент удален.')
    if y == 0:
        print('\nДокумент не найден в базе.')
    print('\nТекущий список документов:')
    inf()
    return
```

```python
def mov_doc():
    """
    Пользователь по команде может переместить документ с полки на полку
    """
    x = input('\nВведите номер документа:\n')
    z = input('\nВведите номер полки:\n')
    y = 0
    if z not in list(directories.keys()):
        print('\nТакой полки не существует. Текущий перечень полок: ' +', '.join(list(directories.keys())))
    else:
        for document in documents:
            if document['number'] == x:
                y = document['number']
                for i,k in directories.items():
                    if x in k:
                        k.remove(x)
                    if i == z:
                        k.append(x)
                print('\nДокумент перемещен.')
        if y == 0:
            print('\nДокумент не найден в базе.')
        print('Текущий список документов:')
        inf()
    return
```
