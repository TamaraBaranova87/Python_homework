### Будем анализировать данные за 2018-2019 гг 

```python
import seaborn as sns
```

```python
sns.pairplot(df[['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy','Freedom to make life choices', 'Generosity', 'Perceptions of corruption']])
```

Уровень счастья больше всего зависит (прямая зависимость) от ВВП на душу населения, соц. защиты и свободы выбора. В более счастливых странах продолжительность жизни выше.

```python
df=df[df['Year']==2019]
top10_countries = df[['Country or region', 'Overall rank']].groupby(by='Country or region').sum().sort_values(by='Overall rank', ascending=False).head(10)
ax = top10_countries.plot(kind='bar')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')
```

```python
Самая несчастливая страна в 2019 г. - Южный Судан.
```

```python
df=df[df['Year']==2019]
top10_countries = df[['Country or region', 'Overall rank']].groupby(by='Country or region').sum().sort_values(by='Overall rank', ascending=True).head(10)
ax = top10_countries.plot(kind='bar')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')
```

```python
Самая счастливая страна в 2019 г. - Финляндия.
```

```python
sns.boxplot(x="Year", y="Freedom to make life choices", 
            data=df)
```

В 2019 г. по сравнению с 2018 г. в целом по всем странам стало меньше свободы выбора.

```python
countries_max = df[(df['Overall rank'] < 11)].head(40)                                            
countries_max
```

```python
happy_countries = df[df.isin(countries_max)].pivot_table(
                        index='Year', 
                        columns='Country or region', 
                        values='Healthy life expectancy', 
                        aggfunc=sum).fillna(0)

happy_countries
```

```python
countries_min = df[(df['Overall rank'] >145)].head(20)                                            
countries_min
```

```python
unhappy_countries = df[df.isin(countries_min)].pivot_table(
                        index='Year', 
                        columns='Country or region', 
                        values='Healthy life expectancy', 
                        aggfunc=sum).fillna(0)

unhappy_countries
```

```python
sns.heatmap(unhappy_countries)
```

```python
sns.heatmap(happy_countries)
```

Как в счастливых, так и в несчастливых странах продолжительность жизни растет, кроме Сирии (видимо, связано с военными действиями).
